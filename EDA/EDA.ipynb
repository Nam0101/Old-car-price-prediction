```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('../data/data.csv')
```

# EDA for old car data

```python
threshold = 0.01  # Ngưỡng 5%
df['brand'] = df['brand'].apply(lambda x: 'Other' if df['brand'].value_counts()[x] / len(df) < threshold else x)
```

```python
df['brand'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# biểu đồ giá trị trung bình của từng hãng xe
plt.figure(figsize=(12, 6))
sns.barplot(x='brand', y='price', data=df)
plt.show()
```

```python
# biểu đồ giá của xe theo năm sản xuất
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='price', data=df)
plt.show()
```

```python
#biểu đồ giá xe theo series
plt.figure(figsize=(12, 6))
sns.barplot(x='series', y='price', data=df)
plt.show()
```

```python
# biểu đồ tỉ lệ engine type
df['engine_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# biểu đồ giá xe theo engine type
plt.figure(figsize=(12, 6))
sns.barplot(x='engine_type', y='price', data=df)
plt.show()
```

```python
#biểu đồ tỉ lệ transmission
df['transmission'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# biểu đồ giá xe theo transmission
plt.figure(figsize=(12, 6))
sns.barplot(x='transmission', y='price', data=df)
plt.show()
```

```python
# biểu đồ tỉ lệ nơi sản xuất
df['assemble_place'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# biểu đồ giá xe theo nơi sản xuất
plt.figure(figsize=(12, 6))
sns.barplot(x='assemble_place', y='price', data=df)
plt.show()
```
